## Module 2 Challenge:

In this assignment, you will create a Python script that analyses the financial records of your company. Inside your starter code, you will find a financial dataset in the budget_data.csv file. This dataset is composed of two columns, Date and Profit/Losses. (Thankfully, your company has rather lax standards for accounting, so the records are simple.)

Your task is to create a Python script that analyses the records to calculate each of the following:

-The total number of months included in the dataset

-The net total amount of Profit/Losses over the entire period

-The average of the changes in Profit/Losses over the entire period

-The greatest increase in profits (date and amount) over the entire period

-The greatest decrease in losses (date and amount) over the entire period

```python
"""Financial Analyses_Module 2 Challenge

This script will use the Pathlib library to set the file path,
use the csv library to read in the file.
"""
# Import libraries
from pathlib import Path
import csv
import os
import shutil
from datetime import datetime

# Set the file path
csvpath = Path("/Users/kiransohi/Documents/Github/Python_Challenge/PyBank/budget_data.csv")

# Initialize dictionary for csvreader
analysis = {}

# Open the csv file as an object
with open(csvpath, "r") as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)
    # Print the header
    #print(csv_header)
    
     # Read each row of data after the header
    for row in csvreader:
        # Print the row
        #print(row)
        
        # Set the 'date' and 'pnl' variables for better readability
        # readability, convert strings to ints for numerical calculations
        date = row[0]
        pnl = row[1]
        
        date, pnl = row
        analysis[date] = pnl
        
        analysis_int = dict((date,int(pnl)) for date, pnl in analysis.items())
    
# Initialize variables
month_count = 0
total_pnl = 0
avg_chg = 0
max_date = 0
min_date = 0
max_value = 0
min_value = 0

# The total number of months included in the dataset.
month_count = len(analysis_int)

# The net total amount of Profit/Losses over the entire period.
total_pnl = sum(analysis_int.values())

# The average of the changes in Profit/Losses over the entire period.
# Initialize lists for subtraction
values = []
dates = []
list_1 = []
list_2 = []

# Create list from keys and values
for i in analysis_int.values():
    values = analysis_int.values()
    values = list(values)

for j in analysis_int.keys():
    dates = analysis_int.keys()
    dates = list(dates)
dates = dates[1:]

# Add century to the dates
s = "20"
dates = [s + x for x in dates]

# Remove the first and last values to subtract the two lists
list_1 = values
list_1 = list_1[:-1]

list_2 = values
list_2 = list_2[1:]

# Use array to complete subtraction and create a new 'subtracted' list
import numpy as np
array1 = np.array(list_2)
array2 = np.array(list_1)
subtracted_array = np.subtract(array1, array2)
subtracted = list(subtracted_array)

# Convert lists of 'date' (key) and 'subtracted' (value) to a new dictionary
pnl_change = {}
for k in dates:
    for v in subtracted:
        pnl_change[k] = v
        subtracted.remove(v)
        break

# Average of the changes in Profit/Losses over the entire period.
avg_chg = sum(pnl_change.values()) / len(pnl_change)

# The greatest increase in profits (date and amount) over the entire period.
max_date = max(pnl_change, key=pnl_change.get)
max_value = max(pnl_change.values())

# The greatest decrease in losses (date and amount) over the entire period.
min_date = min(pnl_change, key=pnl_change.get)
min_value = min(pnl_change.values())

# Parse and Format date for ouput to 'mmm-yyyy'
max_date_n = datetime.strptime(max_date, '%Y-%b').date()
min_date_n = datetime.strptime(min_date, '%Y-%b').date()
max_date_f = max_date_n.strftime("%b-%Y")
min_date_f = min_date_n.strftime("%b-%Y")


# Output
print ("")
print ("")
print (f"Financial Analysis")
print (f"----------------------------")
print (f"Total Months: {month_count}")
print (f"Total:  ${total_pnl}")
print (f"Average Change: ${round(avg_chg, 2)}")
print (f"Greatest Increase in Profits: {max_date_f} $({max_value})")
print (f"Greatest Decrease in Profits: {min_date_f} $({min_value})")
print ("")
print ("")

# Output to Terminal
import sys
sys.stdout = open('/dev/stdout', 'w')


# Output to file name 'Output.txt'
output_path = '/Users/kiransohi/Documents/Github/Python_Challenge/PyBank/output.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    # Write daily_average to the output file, convert to string
    file.write(f"Financial Analysis\n")
    file.write(f"----------------------------\n")
    file.write(f"Total Months: {month_count}\n")
    file.write(f"Total:  ${total_pnl}\n")
    file.write(f"Average Change: ${round(avg_chg, 2)}\n")
    file.write(f"Greatest Increase in Profits: {max_date_f} $({max_value})\n")
    file.write(f"Greatest Decrease in Profits: {min_date_f} $({min_value})\n")
    
# Sources: The following are sources used to code the script above
#
# Source #1:
# Stackoverflow: This source was used (a lot) for dictionary conversion to list 
# and two lists to dictionary, adding century to dates, subtraction of lists, etc.
#
# Source #2:
# docs.python.org: Used for datetime formatting
#
# Source #3:
# www.geeksforgeeks.org:  Used to review the code in application of loops, dicts,
# lists, etc. Great examples of code that shows multiple ways of coding.
```
